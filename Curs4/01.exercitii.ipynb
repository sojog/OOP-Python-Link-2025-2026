# In dictionar cheile sunt unice

```python
johannis = {'cars':2 , 'houses':6}
johannis['cars'] = 3
print(johannis)
```

```python
johannis = {'cars':2 , 'houses':6}
johannis['CARS'] = 3
print(johannis)
```
