```python
## ANIMAL este clasa de baza (din care se mosteneste)
class Animal:
    def __init__(self, nume, greutate, varsta):
        self.nume = nume
        self.greutate = greutate
        self.varsta = varsta 

    def __str__(self):
        return f"{self.nume} are {self.greutate} kg si {self.varsta} ani"

class Mamifer(Animal):
    def naste(self):
        return True


## PISICA este clasa derivata
class Pisica(Mamifer):
    def toarce(self):
        print("pr"*10)


## CAINE este clasa derivata
class Caine(Mamifer):
    def latre(self):
        print("ham" * 3)


        
```

```python
pisica1 = Pisica("Tom", 4, 5)
print(pisica1)
```

```python
pisica1.toarce()
```

```python
caine1 = Caine("Spike", 5, 8)
print(caine1)
```

```python
caine1.latre()
```

```python
pisica1.naste()
```
