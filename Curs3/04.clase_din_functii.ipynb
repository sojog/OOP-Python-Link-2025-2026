```python
vocale = "aeiouAEIOU"

def numara_vocale_consoane(text):
    response = {"consoane":0, "vocale":0}
    for i in text:
        if i.isalpha():
            if i in vocale:
                response["vocale"] += 1
            else:
                response["consoane"] += 1
    return response
```

```python

## CE ARE -> atribute (variabile)
## CE FACE -> metode (functii)

class Procesator:
    ## CE ARE -> atribute (variabile) -> de obiect initializez
    def __init__(self, text):
        self.text = text
        self.vocale = "aeiouAEIOU"

    ## CE FACE -> metoda(functie care are si self)
    def numara_vocale_consoane(self):
        response = {"consoane":0, "vocale":0}
        for i in self.text:
            if i.isalpha():
                if i in self.vocale:
                    response["vocale"] += 1
                else:
                    response["consoane"] += 1
        return response

```

```python
obiect1 = Procesator("Salutare, ce mai faci?" )
obiect1.numara_vocale_consoane()
```

```python
obiect1 = Procesator("Cate exercitii de OOP avem astazi" )
obiect1.numara_vocale_consoane()
```
