```python
vocale = "aeiouAEIOU"
input_string = "Salutare, ce mai faci?" 
```

```python
def numara_vocale_consoane(text):
    response = {"consoane":0, "vocale":0}
    for i in text:
        if i.isalpha():
            if i in vocale:
                response["vocale"] += 1
            else:
                response["consoane"] += 1
    return response
```

```python
for i in input_string:
    if i in vocale:
        print("VOCALA:", i)
    else:
        print("consoana:", i)
```
