# Ex 1

```python
def fun(**a):
    return (min(a)+min(a))/2
print(fun(11,29,51,-31,20))
```

```python
def fun(*a):
    print("a=", a, type(a))
    return (min(a)+min(a))/2
print(fun(11,29,51,-31,20))
```

```python
def fun(*args):
    print("args=", args, type(args))
    return (min(args)+min(args))/2
print(fun(11,29,51,-31,20))
```

```python
def fun(**kargs):
    print("kargs=", kargs, type(kargs))
    
    return (min(kargs)+min(kargs))
print(fun(m=11,n=29,o=51,p=-31,q=20))
```

# Ex 2

```python
def fun(**a, *b):
    return sum(b)/len(b)
print(fun(11,29))
```

```python
def fun(*b, **a):
    return sum(b)/len(b)
print(fun(11,29))
```

```python
def fun(c, *b, **a):
    print("c=", c)
    print("b=", b)
    return sum(b)/len(b)
print(fun(11,29))
```

```python
# Ordinea trebuie sa fie aceasta
def fun(c, *b, **a):
    print("c=", c)
    print("b=", b)
    return sum(b)/len(b)
print(fun(11,29))
```

# Ex 3

```python
def fun(*b, **a):
    return sum(b)/len(b)
print(fun(a=11,b=29))
```

```python
def fun(*b, **a):
    print("a=",a)
    print("b=",b)
    return sum(b)/len(b)
print(fun(x=11,y=29))
```

```python
def fun(*b, **a):
    print("a=",a)
    print("b=",b)
    return sum(b), len(b)
print(fun(a=11,b=29))
```
