# Parametri arbitrari -> *args -> tuple

```python
def suma(*args):
    print(args, type(args))

suma(2, 3)
```

```python
suma(2, 3, 4, 5, 6, 7, 7, 8)
```

```python
suma()
```

```python
def functie(*args, **kargs):
    print(args, type(args))
    print(kargs, type(kargs))
```

```python
functie()
```

```python
functie(32, 231, 321, 321)
```

```python
functie(32, 231, 321, 321, a = 99, b =88, c = 312)
```
